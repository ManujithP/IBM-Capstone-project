```python
import pandas as pd

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv")

# Finding duplicates

df_dup = df[df.duplicated()]

df_dup
```

```python
# Removing duplicates

df.drop_duplicates(inplace = True)

df
```

```python
# Verify if duplicates were actually dropped

df_removed = df[df.duplicated()]

df_removed
```

```python
# Finding Missing Values

# Count total NaN at each column in DataFrame
print(df.isnull().sum())

# Count total NaN in DataFrame 
print('Count of total NaN in DataFrame',df.isnull().sum().sum())
```

```python
# Imputing missing values
# Find the value counts for the column WorkLoc
print(df.WorkLoc.value_counts())
```

```python
# Impute (replace) all the empty rows in the column WorkLoc with the value that you have identified as majority

df["WorkLoc"].fillna("Office", inplace = True)

print(df.WorkLoc.value_counts()) # 32 Office values have replaced NaN values in the WorkLoc column 
```

```python
# Normalizing data
# List out the various categories in the column 'CompFreq'
print(df.CompFreq.value_counts())
```

```python
# Create a new column named 'NormalizedAnnualCompensation'

if df['CompFreq'] == 'Yearly':
    df['NormalizedAnnualCompensation'] = df['CompTotal']
elif df['CompFreq'] == 'Monthly':
    df['NormalizedAnnualCompensation'] = df['CompTotal']*12
elif df['CompFreq'] == 'Weekly':
    df['NormalizedAnnualCompensation'] = df['CompTotal']*52

df['NormalizedAnnualCompensation']
```

```python
import numpy as np

df['NormalizedAnnualCompensation'] = np.where(
    df['CompFreq'] == 'Weekly', df['CompTotal']*52, np.where(
    df['CompFreq'] == 'Monthly', df['CompTotal']*12, df['CompTotal']))

df['NormalizedAnnualCompensation']
```
