```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

print('Dataframe loaded')
```

```python
# view the percentiles
df['ConvertedComp'].describe()
```

```python
# Find out the Inter Quartile Range for the column ConvertedComp
Q_1 = df['ConvertedComp'].quantile(.25)
Q_3 = df['ConvertedComp'].quantile(.75)
IQR = Q_3 - Q_1
print('Q1:', Q_1)
print('Q3:', Q_3)
print('IOR:', IQR)
```

```python
# Find out the upper and lower bounds
Q_1 = df['ConvertedComp'].quantile(.25)
Q_3 = df['ConvertedComp'].quantile(.75)
IQR = Q_3 - Q_1
L_B = Q_1 - (1.5*IQR)
U_B = Q_3 + (1.5*IQR)
print('Lower Bound',L_B)
print('Upper Bound',U_B)
```

```python
df['ConvertedComp'].dtypes
```

```python
# Identify how many outliers are there in the ConvertedComp column
print('Count of outliers',sum((df['ConvertedComp']<L_B) | (df['ConvertedComp']>U_B)))
```

```python
# Create a new dataframe by removing the outliers from the ConvertedComp column
# transform the ConvertedComp column
df['ConvertedComp'] = np.where((df['ConvertedComp']<L_B) | (df['ConvertedComp']>U_B),None,df['ConvertedComp'])
df['ConvertedComp']
```

```python
df['ConvertedComp'].dtypes
```

```python
# Drop all rows with None values in the ConvertedComp column
df.drop(df[df['ConvertedComp'] == None].index, inplace = True)
df.tail(50)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

print('Dataframe loaded')
```

```python
# Find out the upper and lower bounds
Q_1 = df['ConvertedComp'].quantile(.25)
Q_3 = df['ConvertedComp'].quantile(.75)
IQR = Q_3 - Q_1
L_B = Q_1 - (1.5*IQR)
U_B = Q_3 + (1.5*IQR)
print('Lower Bound',L_B)
print('Upper Bound',U_B)

# Create a new dataframe by removing the outliers from the ConvertedComp column
# transform the ConvertedComp column
df['ConvertedComp'] = np.where((df['ConvertedComp']<L_B) | (df['ConvertedComp']>U_B),None,df['ConvertedComp'])
df['ConvertedComp']

# Drop all rows with None values in the ConvertedComp column
df.drop(df[df['ConvertedComp'] == None].index, inplace = True)
df.tail(50)
```

```python
df['ConvertedComp']
```

```python
df['ConvertedComp'].isnull().sum()
```

```python
# take all values(rows) except nan from the df 
df = df[df['ConvertedComp'].notna()]
df['ConvertedComp'].isnull().sum()
```

```python
# Convert ConvertedComp from float64 to int
df[['ConvertedComp']] = df[['ConvertedComp']].astype(int)
df['ConvertedComp'].dtypes
```
